####  write a function that you can apply over the entire DataFrame to verify some of these assumptions. Note that spending the time to write such a script will help you when working with other datasets as well. Concretelly, whether first column is "countries", and entire columns are numeric? Do we have duplicate data?  

```python
#### Переместим life expectancy в начало и соединим датафрейм для анализа
g1800s = pd.concat([gapminder['Life expectancy'],gapminder.loc[:,"1800":'1899']], axis = 1)
g1800s.head()
```

#### Indeed, first column is 'Countries', rest columns are numeric and positive and there is no duplicates. 

```python
def check_null_or_valid(row_data):
    """Function that takes a row of data,
    drops all missing values,
    and checks if all remaining values are greater than or equal to 0
    """
    no_na = row_data.dropna()[1:-1]
    numeric = pd.to_numeric(no_na)
    greater0 = numeric >= 0 # больше нуля
    return greater0

# Check whether the first column is 'Life expectancy'
assert g1800s.columns[0] == 'Life expectancy'
# Check whether the values in the row are valid
assert g1800s.apply(check_null_or_valid, axis =1).all().all()
print(g1800s['Life expectancy'].value_counts()[0]) # [0] самое частое значение
# Check that there is only one instance of each country
assert g1800s['Life expectancy'].value_counts()[0] == 1
    
```

####  Reshaping your data, converting data in tidy format

```python
gapminder_melt = pd.melt(gapminder, id_vars = 'Life expectancy')
gapminder_melt.columns = ['country', 'year', 'life_expectancy']
gapminder_melt.head()
```

#### Review dtypes of gapminder_melt

```python
import numpy as np
gapminder_melt.year = pd.to_numeric(gapminder_melt.year)
assert gapminder_melt.country.dtypes == np.object
assert gapminder_melt.year.dtypes == np.int64
assert gapminder_melt.life_expectancy.dtypes == np.float64
```

```python
gapminder_melt.dtypes
```

#### Since the assert statements did not throw any errors, we can be sure that our columns have the correct data types!

#### next task in the data cleaning process is to look at the 'country' column to see if there are any special or invalid characters you may need to deal with. It is reasonable to assume that country names will contain:
- The set of lower and upper case letters.
- Whitespace between words.
- Periods for any abbreviations.


```python
# Create the series of countries: countries
countries = gapminder_melt.country

# Drop all the duplicates from countries
countries = countries.drop_duplicates()
```

```python
# Write the regular expression: pattern
pattern = '^[A-Za-z\.\s]*$'

# Create the Boolean vector: mask
mask = countries.str.contains(pattern)
mask[40:55]
```

```python
# Invert the mask: mask_inverse
mask_inverse = ~mask

# Subset countries using mask_inverse: invalid_countries
invalid_countries = countries.loc[mask_inverse]

# Print invalid_countries
print(invalid_countries)
```

#### As you can see, not all these country names are actually invalid so maybe the assumptions need to be tweaked a little. However, there certainly are a few cases worth further investigation, such as St. Barth?lemy

#### It's now time to deal with the missing data. There are several strategies for this: You can drop them, fill them in using the mean of the column or row that the missing value is in (also known as imputation), or, if you are dealing with time series data, use a forward fill or backward fill, in which you replace missing values in a column with the most recent known value in the column

#### We need to drop all the rows that have NaN in the life_expectancy column. Before doing so, it would be valuable to use assert statements to confirm that year and country do not have any missing values.

```python
# Assert that country does not contain any missing values
assert pd.notnull(gapminder_melt.country).all()

# Assert that year does not contain any missing values
assert pd.notnull(gapminder_melt.year).all()

# Print the shape of gapminder
print(gapminder_melt.shape)
# Drop the missing values
gapminder_melt = gapminder_melt.dropna(axis = 0, how ='any')

# Print the shape of gapminder after drop na
print(gapminder_melt.shape)
```

#### After dropping the missing values from 'life_expectancy', the number of rows in the DataFrame has gone down from 56203 to 43857. In general, you should avoid dropping too much of your data, but if there is no reasonable way to fill in or impute missing values, then dropping the missing data may be the best solution.

####  investigate how average life expectancy changed over the years. To do this, we need to subset the data by each year, get the life_expectancy column from each subset, and take an average of the values. You can achieve this using the .groupby() method

```python
import matplotlib.pyplot as plt
# Add first subplot
plt.subplot(2, 1, 1) 

# Create a histogram of life_expectancy
gapminder_melt.life_expectancy.plot(kind = 'hist')

# Group gapminder: gapminder_agg
gapminder_agg = gapminder_melt.groupby('year')['life_expectancy'].mean()

# Print the head of gapminder_agg
print(gapminder_agg.head())

# Print the tail of gapminder_agg
print(gapminder_agg.tail())

# Add second subplot
plt.subplot(2, 1, 2)

# Create a line plot of life expectancy per year
gapminder_agg.plot()

# Add title and specify axis labels
plt.title('Life expectancy over the years')
plt.ylabel('Life expectancy')
plt.xlabel('Year')

# Display the plots
plt.tight_layout()
plt.show()

# Save both DataFrames to csv files
gapminder_melt.to_csv('gapminder_melt.csv')
gapminder_agg.to_csv('gapminder_agg.csv')
```

#### Looking at the line plot, it seems like life expectancy has, as expected, increased over the years. Глядя на график, видим, что продолжительность жизни расла с годами, однако, обнаруживается скачок вниз в районе 1920 года. Посмотрим далее.
